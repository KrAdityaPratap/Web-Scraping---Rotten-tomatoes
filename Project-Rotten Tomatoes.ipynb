```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://editorial.rottentomatoes.com/guide/140-essential-action-movies-to-watch-now/"
```

```python
page = requests.get(url)
```

```python
page.status_code
```

```python
page.ok
```

```python
html = page.content
```

```python
soup = BeautifulSoup(html, "lxml")
```

```python
soup
```

```python
all_info = soup.find_all("div", {"class" : "col-sm-18 col-full-xs countdown-item-content"})
```

```python
all_info
```

```python
name_info = [m.find("h2") for m in all_info]
```

```python
movie_name = [m.find("a").string for m in name_info]
```

```python
len(movie_name)
```

```python
release_year = [int(m.find("span", {"class" : "subtle start-year"}).text.strip("()")) for m in name_info]
```

```python
release_year
```

```python
rating = [int(r.find("span", {"class" : "tMeterScore"}).text.strip("%")) for r in name_info]
```

```python
rating
```

```python
all_info[0]
```

```python
critics_consensus = [c.find("div", {"class" : "info critics-consensus"}).text for c in all_info]
```

```python
critics_consensus
```

```python
r = "Critics Consensus: "
```

```python
s = len(r)
```

```python
critics_consensus = [c[s:] for c in critics_consensus]
```

```python
critics_consensus
```

```python
synopsis = [s.find("div", {"class" : "info synopsis"}).text for s in all_info]
```

```python
a = "Synopsis: "
```

```python
b = len(a)
```

```python
synopsis = [l[b:] for l in synopsis]
```

```python
synopsis
```

```python
cast = [c.find("div", {"class" : "info cast"}).text for c in all_info]
```

```python
cast
```

```python
v = "\nStarring: "
```

```python
si = len(v)
```

```python
cast = [c[si:] for c in cast]
```

```python
cast
```

```python
director = [d.find("div", {"class": "info director"}).text for d in all_info]
```

```python
director
```

```python
e = "\nDirected By: "
```

```python
si = len(e)
```

```python
director = [d[si:] for d in director]
```

```python
director
```
